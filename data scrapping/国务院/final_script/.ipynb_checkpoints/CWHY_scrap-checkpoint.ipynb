```python
link_list = menu_df.link
```

```python
total_string_list = []
wrong_list = []

for link in link_list:
    
    URL = link
    
    try:
        response = re.get(URL, HEADER, timeout=10)
        response.encoding = "utf-8"
        text = response.text
        html_file = BeautifulSoup(text,"html.parser")
        content = html_file.find("div",{"class":"content"})
        p_list = content.find_all("p")
        string_list = [p.string for p in p_list if p.string is not None]
        
    except:
        
        wrong_list.append(URL)
        string_list = []
    
    
    total_string = " ".join(string_list)
    
    total_string_list.append(total_string)
```

```python
wrong_list = link_list
```

```python
append_total_string_list = {}
append_wrong_list = []

for link in wrong_list:
    
    URL = link
    
    try:
        response = re.get(URL, HEADER, timeout=10)
        response.encoding = "utf-8"
        text = response.text
        html_file = BeautifulSoup(text,"html.parser")
        content = html_file.find("div",{"class":"pages_content"})
        p_list = content.find_all("p")
        string_list = [p.string for p in p_list if p.string is not None]
        
    except:
        
        append_wrong_list.append(URL)
        string_list = []
    
    
    total_string = " ".join(string_list)
    
    append_total_string_list[URL] = total_string
```

```python
menu_df["content"] = total_string_list
```

```python
for key in append_total_string_list.keys():
    
    menu_df["content"][menu_df.link == key] = append_total_string_list[key]
```

```python
menu_df.to_csv(directory+"GWY_CWHY_content.xls")
```
