```python
import pandas as pd
import os
```

```python
data_directory = "data/output/"
output_directory = "data/output/index/"
```

```python
file_names = os.listdir(data_directory)
```

```python
daily_index = [file for file in file_names if file.endswith("d.csv")]
weekly_index = [file for file in file_names if file.endswith("w.csv")]
monthly_index = [file for file in file_names if file.endswith("m.csv")]
```

```python
file = daily_index[0]
```

```python
df = pd.read_csv(data_directory+file)
```

```python
df["date"] = pd.to_datetime(df["date"])
df = df.set_index("date")
```

```python
daily_index
```

```python
daily_index_name = ["economy","reform","intervention","reform_market","finance","intervention_spending","regulation",
                   "intervention_tax_cut","intervention_support","reform_innovation"]
```

```python
daily_index_list = []

for file in daily_index:
    df = pd.read_csv(data_directory+file)
    df["date"] = pd.to_datetime(df["date"])
    df = df.set_index("date")
    
    daily_index_list.append(df)
```

```python
daily_df = pd.concat(daily_index_list,axis=1)
```

```python
daily_df.columns = daily_index_name
```

```python
weekly_index
```

```python
weekly_index_name = ["economy","reform_market","intervention","reform","intervention_support","intervention_tax_cut",
                    "regulation","intervention_spending","finance","reform_innovation"]
```

```python
weekly_index_list = []

for file in weekly_index:
    df = pd.read_csv(data_directory+file)
    df["date"] = pd.to_datetime(df["date"])
    df = df.set_index("date")
    
    weekly_index_list.append(df)
```

```python
weekly_df = pd.concat(weekly_index_list,axis=1)
```

```python
weekly_df.columns = weekly_index_name
```

```python
weekly_df.plot(subplots=True,figsize=(10,60))
```

```python
monthly_index
```

```python
monthly_index_name = ["reform_innovation","finance","intervention_spending",
                     "regulation","intervention_tax_cut","intervention_support","reform",
                     "intervention","reform_market","economy"]
```

```python
monthly_index_list = []

for file in monthly_index:
    df = pd.read_csv(data_directory+file)
    df["date"] = pd.to_datetime(df["date"])
    df = df.set_index("date")
    
    monthly_index_list.append(df)
```

```python
monthly_df = pd.concat(monthly_index_list,axis=1)
```

```python
monthly_df.columns = monthly_index_name
```

```python
monthly_df.plot(subplots=True,figsize=(10,60))
```
